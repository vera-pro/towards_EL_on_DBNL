```python
import mkwikidata


def get_wikidata_for_author(author_id="mand001"):
    query = """
        select ?wikidata_id ?dbnl_id ?dbnl_url 
        where {
          ?wikidata_id wdt:$prop ?dbnl_id ;
                       wdtn:$prop ?dbnl_url .
          FILTER ( regex(str(?dbnl_url), str(?dbnl_id) , "i" ) ) #alleen de dbnl-url's die horen bij het dbnl-id
        {select ?dbnl_id
        where { ?wikidata_id wdt:$prop ?dbnl_id . # wikidata-id's die een DBNL-auter-ID hebben

        }
        GROUP BY ?dbnl_id
                 }
        FILTER(CONTAINS(?dbnl_id, '$dbnl_value'))
        }
    """
    if author_id[0].isalpha():
        prop = 'P723' # DBNL
    else:
        prop = 'P1006' # Nationale Thesaurus voor Auteursnamen
    result = mkwikidata.run_query(query, params={ "dbnl_value": author_id,"prop":prop })
    return result['results']['bindings'][0]['wikidata_id']['value']
 
```

```python
author_links = []
authors_no_links = []
potential_author_links = [] # if we think it's a place but it's an author

for dbnl_id in tqdm(set(df['id'].to_list())):
    if dbnl_id in authors_wikidata_dict:
        continue
        
    name = dbnl_id.split('0')[0] # if name[0].isalpha() else dbnl_id
    if len(name) == 5:
        print('Looks like {1} ({0}) is a place name, not an author name'.format(dbnl_id, id2name[dbnl_id]))
        try:
            potential_author_links.append((dbnl_id,get_wikidata_for_author(dbnl_id)))
        except:
            print('Right, no Q found')
        continue
    try:
        author_links.append((dbnl_id,get_wikidata_for_author(dbnl_id)))
    except: # not found
        authors_no_links.append(dbnl_id)
```
