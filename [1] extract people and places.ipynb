```python
def strip_numbers(name):
    return "".join([c for c in name if c.isalpha()])
```

```python
strip_numbers('gent_001')
```

```python
def person_name_matches(name, name_id):
    last_name = name.split(' ')[-1]
    return last_name.lower().startswith(strip_numbers(name_id))
```

```python
def place_name_matches(name, name_id): 
    return name.replace(' ','').lower().startswith(strip_numbers(name_id))
```

```python
def look_at_annotations(filename):
    df_same_doc = df[df['filename'] == filename]
    annotations = set(df_same_doc[df_same_doc['type']=='annotatie']['id'])
    names = set(df_same_doc[df_same_doc['type']!='annotatie']['id'])
    return names, annotations
```

```python
for filename in set(df['filename']):
    names, annotations = look_at_annotations(filename)
    if min(len(names),len(annotations)) == 0:
        continue
    print(", ".join(sorted(annotations)))
    print(", ".join(sorted(names)))
    print('=======')
```

```python
def match_annotations(filename):
    df_same_doc = df[df['filename'] == filename]
    annotations = set(df_same_doc[df_same_doc['type']=='annotatie']['id'])
    names = set(df_same_doc[df_same_doc['type']!='annotatie']['id'])
    res = []
    for name in names:
        for annotation in annotations:
            if person_name_matches(name, annotation) or place_name_matches(name, annotation):
                res.append((name, annotation))
    return res
```

```python
matches_per_doc = {
    filename: match_annotations(filename)
    for filename in set(df['filename'])
}
```

```python
matches_per_doc_exist = {
    filename: matches_per_doc[filename]
    for filename in set(df['filename']) if matches_per_doc[filename]

}
```

```python
def get_text(filename):
    path = 'data/xml_pd/'+filename
    return open(path,'r').read()
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup()
```

```python
import re
records = []

for filename in matches_per_doc_exist.keys():
    soup = BeautifulSoup(get_text(filename))
    text = soup.get_text()
    
    for name, name_id in matches_per_doc_exist[filename]:
        for m in re.finditer(name, text):
            start = m.start()
            records.append({'name': name,
                           'id': name_id,
                           'left_context': text[max(0, start-140):start],
                           'right_context': text[start+len(name):min(len(text), start+len(name)+140)],
                           'filename': filename})
```

```python
df_dataset = pd.DataFrame.from_records(records)
df_dataset.to_csv('DBNL_links.tsv',sep='\t')
df_dataset
```

```python
def is_person(name):
    return name[-3:].isnumeric() and name[:-3].isalpha()

def is_place(name):
    return name[-2:].isnumeric() and name[:-2].isalpha()
```
